# Evaluation demo
This notebook reproduces the evaluation framework of the Formula 1 Retrieval-Augmented Generation (RAG) chatbot.


```python
import pandas as pd

from evaluation import (
    evaluate_retrieval_metrics,
    evaluate_generation_metrics,
    evaluate_latency,
    evaluate_sources_and_citations
)
```

Retrieval metrics

```python
retrieval_results = evaluate_retrieval_metrics(top_k=5)

retrieval_df = pd.DataFrame(retrieval_results)

retrieval_df
```

```python
print(
    f"Average Precision@5: "
    f"{retrieval_df['precision@5'].mean():.3f}"
)

print(
    f"Average Recall@5: "
    f"{retrieval_df['recall@5'].mean():.3f}"
)

print(
    f"Average MRR: "
    f"{retrieval_df['mrr'].mean():.3f}"
)
```

Generation metrics

```python
generation_results = evaluate_generation_metrics(top_k=5)

generation_df = pd.DataFrame(generation_results)

generation_df
```

Latency evaluation

```python
latency_results = evaluate_latency(top_k=5)

latency_df = pd.DataFrame(latency_results)

latency_df
```

```python
print(
    f"Average Retrieval Latency: "
    f"{latency_df['retrieval_latency_seconds'].mean():.3f} seconds"
)

print(
    f"Average Generation Latency: "
    f"{latency_df['generation_latency_seconds'].mean():.3f} seconds"
)
```

Sources evaluation

```python
citation_results = evaluate_sources_and_citations(top_k=5)

citation_df = pd.DataFrame(citation_results)

citation_df
```

```python
print(
    f"Percentage of answers containing citations: "
    f"{citation_df['has_sources'].mean()*100:.1f}%"
)

print(
    f"Percentage of answers containing expected citations: "
    f"{citation_df['contains_expected_source'].mean()*100:.1f}%"
)
```
